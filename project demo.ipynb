```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import seaborn as sns
from scipy.sparse import csr_matrix
```

```python
movies = pd.read_csv('./archive/movies.csv')
ratings = pd.read_csv('./archive/ratings.csv')
```

```python
if movies.isnull().values.any():
    print('Nan in Movie set')
else:
    print('All entries are not null')

if ratings.isnull().values.any():
    print('Nan in Movie set')
else:
    print('All entries are not null')
    
```

```python
strategy1 = ratings.pivot(index='movieId',columns='userId',values='rating')
strategy1.fillna(0,inplace=True)
print(strategy1.shape)
strategy1
```

```python
all_users_count = ratings.groupby('movieId')['rating'].agg('count')
all_movies_count = ratings.groupby('userId')['rating'].agg('count')
strategy1 = strategy1.loc[all_users_count[all_users_count > 10].index,:]
strategy1 = strategy1.loc[:,all_movies_count[all_movies_count > 40].index]
csr_data = csr_matrix(strategy1.values)
strategy1.reset_index(inplace=True)
csr_data
```
